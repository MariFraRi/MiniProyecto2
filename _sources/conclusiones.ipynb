## **Análisis y Reporte Final**

### **Comparación de modelos y valores óptimos de regularización**
Se entrenaron y evaluaron distintos modelos de regresión y clasificación:

- **Regresión**  
  - Ridge: mejor desempeño (R² ≈ 0.66, MAE ≈ 5,242, RMSE ≈ 8,225).  
  - Lasso: desempeño más bajo (R² ≈ 0.46), indicando una regularización excesiva.  
  - Valores óptimos de regularización (`alpha`):
    - Ridge → α óptimo ≈ 1  
    - Lasso → α óptimo ≈ 0.1  

- **Clasificación (Regresión Logística):**  
  - Accuracy, Precision, Recall, y F1 ≈ 0.86  
  - AUC = 0.938 → excelente poder de discriminación.  
  - El parámetro de regularización (C inverso a λ) mostró que valores intermedios logran el mejor equilibrio.

---

### **Variables que afectan más la demanda y el precio**
- **Precio (regresión):**  
  - Variables con mayor impacto: `anio`, `odometro`, `fabricante`, `tipo`.  
  - La variable `anio` (edad del auto) se relaciona fuertemente con un mayor precio.  
  - El `odometro` (kilometraje) tiene un efecto negativo sobre el precio.  

- **Demanda (clasificación):**  
  - Variables más influyentes: `precio`, `anio`, `tipo`, `traccion`.  
  - Autos recientes y con menor kilometraje tienen mayor probabilidad de estar en alta demanda.  
  - Algunas categorías de `tipo` (ej. SUV) concentran más interés.

---

### **Efecto de la regularización en la capacidad de generalización**
- La regularización (Ridge y Lasso) evitó el sobreajuste al penalizar coeficientes grandes.  
- **Ridge** mejoró la estabilidad y redujo el error en test respecto a la regresión lineal básica.  
- **Lasso** eliminó algunas variables irrelevantes, aunque sacrificando capacidad predictiva en este dataset.  
- En clasificación, la **regularización L2** en la Regresión Logística permitió mantener un buen equilibrio entre precisión y recall.
